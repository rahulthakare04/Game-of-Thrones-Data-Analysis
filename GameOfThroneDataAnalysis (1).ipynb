```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
```

load the csv


```python
url="Game_of_Thrones.csv"
df=pandas.read_csv(url)
```

Delete some col the we Don't need
1.  Original Air Date
2.  Filming Duration
3.  Ordered
4.  Synopsis



```python
df=df.drop(['Original Air Date','Filming Duration','Ordered','Synopsis'],axis=1)
```

Number of ep per season

```python
season_ep_count=df.groupby('Season').size()
print('episodes per season:',season_ep_count)
```

Viewership tends over the siason

```python
sns.lineplot(x=df['Season'],y=df['U.S. Viewers (Millions)'])
plt.title("Viewership Over Seasons")
plt.xlabel('Season')
plt.ylabel('U.S. Viewership (Millions)')
plt.show()
```

Top-reted and lowest-reted episodes

```python
top_reated=df.loc[df['IMDb Rating'].idxmax()]
lowest_reated=df.loc[df['IMDb Rating'].idxmin()]
print('highest rated episode:\n',top_reated[['Title of the Episode','IMDb Rating']])
print('lowest rated episode:\n',lowest_reated[['Title of the Episode','IMDb Rating']])
```

top 5 director who direct the most episode

```python
top_directer=df['Directed by'].value_counts().head(5)
print(top_directer)
```

top 5 writers who write the most episode

```python
top_writers=df['Written by'].value_counts().head(5)
print(top_writers)
```

top 5 Cinematographer

```python
top_Cinematograper=df['Cinematography by'].value_counts().head(5)
print(top_Cinematograper)
```

top 5 editer

```python
top_editer=df['Editing by'].value_counts().head(5)
print(top_editer)
```

seasonal trends
Average imdb rating per season

```python
avg_imdb_rating_per_season=df.groupby('Season')['IMDb Rating'].mean()
plt.figure(figsize=(10,5))
sns.barplot(x=avg_imdb_rating_per_season.index,y=avg_imdb_rating_per_season.values)
plt.title('Average IMDb Rating Per Season')
plt.xlabel('Season')
plt.ylabel('Average IMDb Rating')
plt.show()
```

seasonal trends
Average Rotten Tomatoes Rating season

```python
avg_Rotten_Tomatoes_rating_per_season=df.groupby('Season')['Rotten Tomatoes Rating (Percentage)'].mean()
plt.figure(figsize=(10,5))
sns.barplot(x=avg_Rotten_Tomatoes_rating_per_season.index,y=avg_Rotten_Tomatoes_rating_per_season.values)
plt.title('Average Rotten Tomatoes Rating (Percentage) Per Season')
plt.xlabel('Season')
plt.ylabel('Average Rotten Tomatoes Rating (Percentage)')
plt.show()
```

seasonal trends
Average Metacritic Ratings season

```python
avg_Metacritic_Ratings_rating_per_season=df.groupby('Season')['Metacritic Ratings'].mean()
plt.figure(figsize=(10,5))
sns.barplot(x=avg_Metacritic_Ratings_rating_per_season.index,y=avg_Metacritic_Ratings_rating_per_season.values)
plt.title('Average Metacritic Ratings Per Season')
plt.xlabel('Season')
plt.ylabel('Average Metacritic Ratings')
plt.show()
```
